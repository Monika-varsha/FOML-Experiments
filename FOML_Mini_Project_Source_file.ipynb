```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score

# Load your dataset
data = pd.read_csv('all_season_batting_card.csv')  # Replace with your dataset

# Convert any non-numeric values (like '-') to NaN
columns_to_check = ['runs', 'ballsFaced', 'strikeRate']  # Update this list based on your features
for col in columns_to_check:
    data[col] = pd.to_numeric(data[col], errors='coerce')  # 'coerce' converts invalid parsing to NaN

# Select batting score-related features and target variable
features = ['runs', 'ballsFaced', 'strikeRate']

# Assuming we have the final scores for both teams
data['match_outcome'] = (data['runs'] > data['runs'].shift(-1))  # Example of outcome based on runs
data['match_outcome'] = data['match_outcome'].map({True: 1, False: 0})  # 1 = win, 0 = loss

target = 'match_outcome'

# Handle missing values in the features (drop rows with NaN in the selected columns)
data = data.dropna(subset=features)
data
# Split data into training and testing sets
X = data[features]
y = data[target]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train a model
model_df = RandomForestClassifier()
model_df.fit(X_train, y_train)

# Make predictions
predictions = model_df.predict(X_test)

# Evaluate the model
print(classification_report(y_test, predictions))
print("Accuracy:", accuracy_score(y_test, predictions))
```
